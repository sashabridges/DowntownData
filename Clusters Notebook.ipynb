```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# insert data
df = pd.read_csv('CSVs/9 columns KCPD_crime_data_2014.csv')
df = df.dropna()
df.head
```

```python
df = df[["Zip Code", "Offense"]]
df.to_csv("2014cluster.csv")
```

```python
# plot data
plt.scatter(df["Zip Code"], df["Offense"])
```

```python
# Use n_clusters=4 as the k value
# We can see from the plot above that there are 4 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
# Fit the model to the data
kmeans.fit(df)
```

```python
# Predict the clusters
predicted_clusters = kmeans.predict(df)
```

```python
# Plot the predicted clusters to see if the model predicted the correct clusters
# This is visual validation that the model was trained correctly.
plt.scatter(df["Zip Code"], df["Offense"], c=predicted_clusters, s=50, cmap='viridis')
plt.savefig('2014_clusters.png')
```
